## 📊 4. Gráfico 1 — Série Temporal

**Pergunta:** *Como a prevalência de depressão no Brasil evoluiu ao longo do tempo?*  
**Gráfico escolhido:** Linha com área preenchida — ideal para mostrar tendência temporal contínua.

```python
fig, ax = plt.subplots(figsize=(12, 6), facecolor=BG)
ax.set_facecolor(BG)

# Área preenchida
ax.fill_between(df_depressao['ano'], df_depressao['prevalencia_depressao'],
                alpha=0.15, color=PURPLE)

# Linha principal
ax.plot(df_depressao['ano'], df_depressao['prevalencia_depressao'],
        color=PURPLE, linewidth=2.5, marker='o', markersize=5,
        markerfacecolor=MAGENTA, label='Brasil')

# Destaque ponto máximo (cor intencional)
max_row = df_depressao.loc[df_depressao['prevalencia_depressao'].idxmax()]
ax.annotate(f"{max_row['prevalencia_depressao']:.2f}% (pico)",
            xy=(max_row['ano'], max_row['prevalencia_depressao']),
            xytext=(max_row['ano'] - 5, max_row['prevalencia_depressao'] + 0.06),
            fontsize=9, color=MAGENTA, fontweight='bold',
            arrowprops=dict(arrowstyle='->', color=MAGENTA, lw=1.2))

# Título narrativo (não descritivo)
ax.set_title('Crescimento contínuo da depressão no Brasil (1990–2019)',
             fontsize=14, fontweight='bold', color=PURPLE, pad=16)
ax.set_xlabel('Ano', fontsize=11, color=GRAY)
ax.set_ylabel('Prevalência (%)', fontsize=11, color=GRAY)
ax.tick_params(colors=GRAY)
ax.spines[['top','right']].set_visible(False)
ax.spines[['left','bottom']].set_color('#CCCCCC')
ax.grid(axis='y', linestyle='--', alpha=0.4)
ax.set_xlim(1989, 2020)

plt.tight_layout()
plt.savefig('grafico1_serie_temporal.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Gráfico 1 salvo como grafico1_serie_temporal.png')
```

### 💡 Insights — Gráfico 1

| | |
|---|---|
| **O dado** | A prevalência de depressão no Brasil cresceu de **3,50% em 1990** para **4,37% em 2019** — aumento de 24,9% em 30 anos. |
| **O porquê** | O crescimento é gradual e contínuo, sem reversões, indicando causas estruturais: urbanização, estresse crônico e acesso limitado a tratamento. |
| **O insight acionável** | A SAPA deve priorizar o usuário entre 25–45 anos, faixa de maior exposição ao ambiente de alta pressão do mercado de trabalho. |

## 📊 5. Gráfico 2 — Boxplot por Faixa de Renda

**Pergunta:** *Como se distribui o gasto com saúde mental entre diferentes faixas de renda?*  
**Gráfico escolhido:** Boxplot — ideal para mostrar mediana, dispersão e outliers por categoria.

```python
# Ordenar faixas de renda de forma lógica
ordem_renda = ['Até 2 SM', '2-5 SM', '5-10 SM', '10-20 SM', 'Acima de 20 SM']
df_despesas['faixa_renda_familiar'] = pd.Categorical(
    df_despesas['faixa_renda_familiar'], categories=ordem_renda, ordered=True)
df_sorted = df_despesas.sort_values('faixa_renda_familiar')

fig, ax = plt.subplots(figsize=(12, 6), facecolor=BG)
ax.set_facecolor(BG)

palette = [PURPLE, LIGHT_P, MAGENTA, '#7B4BB0', '#E06EB0']
sns.boxplot(data=df_sorted,
            x='faixa_renda_familiar',
            y='pct_saude_mental',
            palette=palette,
            ax=ax,
            width=0.55,
            linewidth=1.2,
            flierprops=dict(marker='o', markersize=3.5,
                            color=GRAY, alpha=0.5))

ax.set_title('Renda alta ≠ mais proteção: desigualdade no gasto com saúde mental',
             fontsize=14, fontweight='bold', color=PURPLE, pad=16)
ax.set_xlabel('Faixa de Renda Familiar', fontsize=11, color=GRAY)
ax.set_ylabel('% da despesa total de saúde em saúde mental', fontsize=11, color=GRAY)
ax.tick_params(colors=GRAY)
ax.spines[['top','right']].set_visible(False)
ax.spines[['left','bottom']].set_color('#CCCCCC')
ax.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.savefig('grafico2_boxplot_renda.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Gráfico 2 salvo como grafico2_boxplot_renda.png')
```

### 💡 Insights — Gráfico 2

| | |
|---|---|
| **O dado** | A faixa 'Acima de 20 SM' apresenta mediana de % gasto com saúde mental similar ou inferior à faixa 'Até 2 SM', com alta variância nos dois extremos. |
| **O porquê** | Renda alta amplia acesso a tratamentos mais caros, mas não elimina o transtorno. A dispersão elevada indica que muitas famílias de alta renda ignoram completamente esse gasto. |
| **O insight acionável** | A SAPA deve ser acessível financeiramente — o mercado de baixa-média renda é tão afetado quanto o de alta renda e carece de soluções digitais de autoconhecimento. |

## 📊 6. Gráfico 3 — Barras Horizontais (Top 10 CID)

**Pergunta:** *Quais diagnósticos psiquiátricos mais motivam internações no Centro-Oeste?*  
**Gráfico escolhido:** Barras horizontais — ideal para comparação categórica com rótulos longos.

```python
# Mapeamento CID → descrição legível
cid_map = {
    'F322': 'Depressão grave s/ psicose (F322)',
    'F200': 'Esquizofrenia paranoide (F200)',
    'F329': 'Depressão não especif. (F329)',
    'F311': 'Transt. bipolar — ep. maníaco (F311)',
    'F208': 'Outras esquizofrenias (F208)',
    'F312': 'Transt. bipolar — ep. depressivo (F312)',
    'F100': 'Transt. por uso de álcool (F100)',
    'F500': 'Anorexia nervosa (F500)',
    'F068': 'Outros transt. mentais orgânicos (F068)',
    'F318': 'Outros transt. bipolares (F318)',
}

top10 = (df_internacoes['cid_diagnostico']
         .value_counts()
         .head(10)
         .reset_index())
top10.columns = ['cid', 'n_internacoes']
top10['label'] = top10['cid'].map(cid_map).fillna(top10['cid'])
top10 = top10.sort_values('n_internacoes')  # crescente p/ barras horizontais

fig, ax = plt.subplots(figsize=(12, 7), facecolor=BG)
ax.set_facecolor(BG)

# Cor intencional: destaque no 1º lugar
colors_bars = [MAGENTA if i == len(top10)-1 else PURPLE
               for i in range(len(top10))]

bars = ax.barh(top10['label'], top10['n_internacoes'],
               color=colors_bars, height=0.65)

# Rótulos nas barras
for bar in bars:
    w = bar.get_width()
    ax.text(w + 4, bar.get_y() + bar.get_height()/2,
            f'{int(w)}', va='center', fontsize=9, color=GRAY)

ax.set_title('Depressão lidera internações em saúde mental no Centro-Oeste (2025)',
             fontsize=14, fontweight='bold', color=PURPLE, pad=16)
ax.set_xlabel('Número de internações', fontsize=11, color=GRAY)
ax.tick_params(colors=GRAY, labelsize=9)
ax.spines[['top','right','left']].set_visible(False)
ax.spines['bottom'].set_color('#CCCCCC')
ax.grid(axis='x', linestyle='--', alpha=0.4)
ax.set_xlim(0, top10['n_internacoes'].max() * 1.15)

legenda = [mpatches.Patch(color=MAGENTA, label='1º lugar'),
           mpatches.Patch(color=PURPLE,  label='Demais diagnósticos')]
ax.legend(handles=legenda, fontsize=9, framealpha=0, loc='lower right')

plt.tight_layout()
plt.savefig('grafico3_barras_cid.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Gráfico 3 salvo como grafico3_barras_cid.png')
```

### 💡 Insights — Gráfico 3

| | |
|---|---|
| **O dado** | Depressão grave sem psicose (F322) lidera com **527 internações** — quase o dobro do 2º colocado (esquizofrenia paranoide, 292). |
| **O porquê** | Transtornos depressivos respondem por mais de 30% de todas as internações psiquiátricas analisadas, confirmando a depressão como principal problema de saúde mental no Brasil Central. |
| **O insight acionável** | O foco inicial da SAPA em rastreamento de humor e sinais de depressão é tecnicamente correto e epidemiologicamente justificado pelos dados reais do SUS. |
